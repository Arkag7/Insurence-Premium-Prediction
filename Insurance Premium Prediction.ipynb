```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
```

```python
X=df.drop('expenses',axis=1)
y=df['expenses']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
model_lr=LinearRegression()
model_lr.fit(X_train,y_train)
model_lr.score(X_test,y_test)
```

```python
model_knn=KNeighborsRegressor()
model_knn.fit(X_train,y_train)
model_knn.score(X_test,y_test)
```

```python
model_rand=RandomForestRegressor()
model_rand.fit(X_train,y_train)
model_rand.score(X_test,y_test)
```

```python
model_ada=AdaBoostRegressor()
model_ada.fit(X_train,y_train)
model_ada.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
y_pred=model_rand.predict(X_test)
```

```python
y_pred
```

```python
print("r2_score",r2_score(y_test,y_pred))
print("mean_absolute_error",mean_absolute_error(y_test,y_pred))
print("mean_squared_error",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {"n_estimators":[100,200,300],
          "criterion":["mse","mae"],
          "max_depth":[None,1,2,5],
          "min_samples_split":[2,4]}
```

```python
model=RandomForestRegressor()
```

```python
model_grid=GridSearchCV(estimator=model,param_grid=params,scoring='r2',n_jobs=-1)
```

```python
model_grid.fit(X_train,y_train)
```

```python
params = {"n_estimators":[100,200,300],
          "criterion":["mse","mae"],
          "max_depth":[None,1,2,3,5],
          "min_samples_split":[2,4]}
```

```python
model_grid=GridSearchCV(estimator=model,param_grid=params,scoring='r2',n_jobs=1,cv=5)
```

```python
model_grid.fit(X_train,y_train)
```

```python
print(model_grid.score(X_test,y_test))
```
